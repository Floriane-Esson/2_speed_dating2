## Cleaning dataset
Create new dataset with only variables of 1st and 2nd date

```python
# create list with variables corresponding to the 3rd date with help of "speed dating data key.doc"
column_to_drop = ["date_3","numdat_3","num_in_3","attr1_3","sinc1_3","intel1_3","fun1_3","amb1_3","shar1_3","attr7_3","sinc7_3","intel7_3","fun7_3","amb7_3","shar7_3","attr4_3","sinc4_3","intel4_3","fun4_3","amb4_3","shar4_3","attr2_3","sinc2_3","intel2_3","fun2_3","amb2_3","shar2_3","attr3_3","sinc3_3","intel3_3","fun3_3","amb3_3","attr5_3","sinc5_3","intel5_3","fun5_3","amb5_3"]
len(column_to_drop)
```

```python
keepcolumns = [col for col in speed_dating.columns if col not in column_to_drop]
len(keepcolumns)
```

```python
speed_dating2 = speed_dating.loc[:,keepcolumns]
```

```python
speed_dating2.shape
```

```python
print("Percentage of missing values on speed_dating2 : ")
display((((speed_dating2.isna().sum())/(speed_dating2.shape[0]))*100).sort_values(ascending=False))
```

```python
missingvalues = (((speed_dating2.isna().sum())/(speed_dating2.shape[0]))*100).sort_values(ascending=False)
missingvalues[missingvalues>60]
```

```python
# remove variables with percentage of missingvalues below 60%
speed_dating2 = speed_dating2.loc[:,missingvalues<60]
speed_dating2.shape
```

```python
# How many match on dataset (Match = 1 / No match = 0)
print("Number of match :")
print(speed_dating2.match.value_counts())
print()
print("Percentage of match :")
print(speed_dating2.match.value_counts()/speed_dating2.match.count()*100)
```

```python
#distribution of the dataset
print("Distribution of gender in dataset :")
print(speed_dating2.gender.value_counts())
print()
print("Distribution of condition in dataset : ")
print("1=limited choice / 2=extensive choice")
print(speed_dating2.condtn.value_counts())
```

# Part II : First visualisations & relationships between variables

Let's try to produce our first visualisations with Seaborn. Based on the exploration you realized, try to find good representations of the dataset that allow to summarize some statistics as well as relationships between variables.

```python
# Create selection of variables to visualise correlation mapping
column_sel = ["match","gender","samerace","age_o","race_o", "int_corr","attr_o","dec_o", "age", "field", "race", "imprelig", "from", "goal", "go_out", "career", "exphappy", "attr1_1", "sinc1_1", "intel1_1", "fun1_1", "amb1_1", "shar1_1","match_es", "condtn"]
sdsel = speed_dating2.loc[:, column_sel]
sdsel.shape
```

```python
# visualise heatmap for correlation between variables in selection
corr = sdsel.corr()
fig1, axes = plt.subplots(figsize=(20,7))
sns.heatmap(corr,annot=True,ax=axes)
plt.show()
```

```python
fig5 = go.Figure()
fig5 = px.histogram(speed_dating, x="match", color="gender", nbins =2, text_auto=True)
fig5.update_layout(bargap=0.2, barmode="group", title = go.layout.Title(text = "Match vs Gender", x = 0.5))
fig5.show()
```

```python
fig6 = go.Figure()
fig6 = px.histogram(speed_dating, x="age", color="match", nbins=10)
fig6.update_layout(
        title = go.layout.Title(text = "Age vs Match", x = 0.5))
fig6.show()
```
